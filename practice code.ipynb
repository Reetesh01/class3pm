```python
import random
n = int(input("enter number of rounds = "))
roll = 0

def dice():
    dice_roll = random.randint(1,6)
    return dice_roll

player1 = input('enter name of player 1 = ')
player2 = input('enter name of player 2 = ')
player1wins = 0
player2wins = 0

for i in range(n):
    roll = roll + 1
    print('ROUND: ',roll)
    
    p1 = dice()
    print(player1 , 'rolls :', p1)
    p2 = dice()
    print(player2 , 'rolls :', p2)
    
    if p1 == p2:
        print('DRAW!!', p1,'\n')
        
    elif p1 > p2:
        player1wins = player1wins + 1
        print(player1,'wins by ',p1)
        print(player1, "won round",player1wins,'\n')
        
    else:
        player2wins = player2wins + 1
        print(player2,"wins by ",p2)
        print(player2, "won round",player2wins,'\n')
        
if player1wins == player2wins:
    print("Well it's a DRAW dare to play again!!")
elif player1wins > player2wins:
    print(player1,"wins by ",player1wins,"rounds!!")
else:
    print(player2, 'wins by ', player2wins, "rounds!!")
```

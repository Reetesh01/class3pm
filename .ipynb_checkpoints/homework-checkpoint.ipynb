```python
def add(a, b):
    return a+b
def sub(a, b):
    return a-b
def mul(a, b):
    return a*b
def divd(a, b):
    return a/b

print("Please select an operation:")
print("1.ADD ")
print("2.SUBSTRACT ")
print("3.MULTIPLY ")
print("4.DIVIDE ")

while True:
    choice = input("Select opearation 1|2|3|4 =")
    if choice in ('1', '2', '3', '4'):
        num1 = float(input("Enter first number = "))
        num2 = float(input("Enter second number = "))
        
        if choice == '1':
            print(num1, '+', num2, '=', num1+num2)
        elif choice == '2':
            print(num1, '-', num2, '=', num1-num2)
        elif choice == '3':
            print(num1, 'x', num2, '=', num1*num2)
        elif choice == '4':
            print(num1, '/', num2, '=', num1/num2)
        
        next_op = input('Do you want to continue next calculation ? yes/no : ')
        if next_op == 'no':
            break
    else:
        print("Invalid Entry!!")
```

```python
class A:
    def __init__(self,name):
        self.name = name
    def name(self):    
        return f'Input Names are = {self.name}'
    
class B:
    def __init__(self,age):
        self.age = age
    def age(self):
        return f'Input Ages are = {self.age}'

class C(A,B):
    def __init__(self,name,age):
        A.__init__(self,name)
        B.__init__(self,age)
        print(super().name())
        print(super().age())
        
    def info(self):
        print(f"name = {self.name}, age = {self.age}")
        
obj = C(['Ram',34],['shyam',24])
obj.info()
```

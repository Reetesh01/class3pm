```python
s = str()
n = int(input('enter n = '))
for i in range(n):
    name = input("enter name = ")
    phone = int(input("enter phone number = "))
    info = f"{name} {phone}\n"
    s = s+info
    
print(s)

b = open('dataas.txt', 'w')
b.write(s)
b.close()
```

```python
s = str()
n = int(input('enter n = '))
for i in range(n):
    name = input("enter name = ")
    phone = int(input("enter phone number = "))
    info = f"{name},{phone}\n"
    s = s+info
    
print(s)

b = open('dataas.csv', 'w')
b.write(s)
b.close()
```

```python
s = str()
n = int(input('enter n = '))
for i in range(n):
    name = input("enter name = ")
    phone = int(input("enter phone number = "))
    info = f"{name},{phone}\n"
    s = s+info
    
print(s)

b = open('dataas.csv', 'w')
b.write("Name,Phone\n")
b.write(s)
b.close()
```

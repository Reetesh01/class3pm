```python
def cal():
    l =10
    b = 5
    h = 2
    a = l*b
    v = a*h
    return a,v
area,volume = cal()
print(area)
print(volume)
```

```python
s = 0
n = int(input ('enter n = '))
def summ():
    global s
    for i in range(n):
        x = int(input('enter x = '))
        s = s+x

summ()
print(s)
```

```python
# Decorator
```

```python
def outer_function(fun):
    
    def inner_function():
        x = fun()
        
    return inner_function()

@outer_function
def hello():
    print('Hello World')
```

```python
def string(function):
    def wrapper():
        fun = function()
        print(fun)
        x = fun.lower()
        return x
    return wrapper

@string
def info():
    return 'Hello World'
info()
```

```python
def hello(function):
    def world():
        x = function()
        y = x.upper()
        return y
    return world


@hello
def info():
    return 'Ram and Shyam'
info()
```

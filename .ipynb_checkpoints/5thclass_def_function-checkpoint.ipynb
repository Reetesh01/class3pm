```python
# # #function
# # function are pre-defined values which can be recalled

# # syntax
# def<function_name()>:
#     operation:
        
# <function_name>
```

```python
def hello():
    print("Hello World")
    
hello()
```

```python
#the variable inside def function are local variable
def cal():
    l = int(input("enter the value  of l = "))
    b = int(input("enter the value of b = "))
    a = l*b
    print(a)

    
    
cal()
```

```python
cal()
```

```python
#the variables outside of def function are called global variablers

l = int(input("enter the value  of l = ")) # global variables
b = int(input("enter the value of b = ")) #global variables

def cal():
    a = l*b #local variables 
    print(a)

    
    
cal()
```

```python
def cal():
    l = int(input("enter the value  of l = "))
    b = int(input("enter the value of b = "))
    a = l*b
    print(a)
n = int(input("enter n= "))
for i in range(n):
    cal()
```

```python
#function with arguments

def info(x):
    print(x)

info("Hello World")
```

```python
def cal(l, b):
    a = l*b
    print(a)
cal(10, 5)
    
```

```python
def cal(l, b):
    a = l*b
    print(a)
l = int(input("enter l = "))
b = int(input("enter b = "))

cal(l, b)
```

```python
#variables to carry and in function can be different
def cal(x, y):
    a = x*y
    print(a)
l = int(input("enter l = "))
b = int(input("enter b = "))

cal(l, b)
```

```python
def language(lan = "python"):
    print(lan)
    
language("c")
language("java")
language("c++")
language()
```

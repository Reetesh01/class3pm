```python
class A:
    def __init__(self,add):
        self.add = add
    def address(self):
        print(f"I'm form {self.add}")

class B(A):
    def __init__(self,age,add):
        self.age = age
        A.__init__(self,add)
    def age(self):
        print(f"I am from {self.add}. I am {self.age}")

class C(B):
    def __init__(self,name,age,add):
        self.name = name
        B.__init__(self,age,add)
    def info(self):
        print(f"Hello World Im {self.name},I am from {self.add}.I am {self.age} years old.")

obj = C('Ram',34,'Kathmandu')
obj.info()
```

```python
class A:
    def __init__(self,add):
        self.add = add
    def address(self):
        print(f"I'm form {self.add}")

class B:
    def __init__(self,age):
        self.age = age
    def age(self):
        print(f"I am {self.age}")

class C(A,B):
    def __init__(self,name,age,add):
        self.name = name
        B.__init__(self,age)
        A.__init__(self,add)
        
    def info(self):
        print(f"Hello World Im {self.name},I am from {self.add}.I am {self.age} years old.")

obj = C('Ram',34,'Kathmandu')
obj.info()
```

```python
# protected member

class info:
    def __init__(self,name,age,add):
        self._name = name
        self._age = age
        self._add = add
obj = info('Ram', 34,'Kathmandu')
print(obj._name)
print(obj._age)
print(obj._add)
print(type(info))
```

```python
class A:
    def __init__(self,name,age,add):
        self.name = name
        self._age = age
        self.__add = add
    def a(self):
        print(f'Hello my name is {self.name}.i am {self._age} years old. i am from {self.__add}')

class B(A):
    def info(self):
        print(f'Hello my name is {self.name}.i am {self._age} years old.')

name = input('enter name = ')
age = int(input("enter age = "))
add = input('enter address = ')
        
obj = B(name,age,add)
obj.info()
obj.a()
```

```python
class A:
    def __init__(self,add):
        self._add = add
    def address(self):
        print(f"I'm form {self.add}")

class B:
    def __init__(self,age):
        self._age = age
    def age(self):
        print(f"I am {self.age}")

class C(A,B):
    def __init__(self,name,age,add):
        self.__name = name
        B.__init__(self,age)
        A.__init__(self,add)
        
    def info(self):
        print(f"Hello World Im {self.__name},I am from {self._add}.I am {self._age} years old.")

obj = C('Ram',34,'Kathmandu')
obj.info()
```

```python
u = {'alex','rem','jobs', 'sabin', 'bill', 'sabita', 'sita', 'ram', 'john', 'hari', 'shyam'}

apple = {'ram','shyam','hari','sita','sabin','jobs','sabin'}
ms = {'bill', 'sabita','hari','shyam','john','sita'}
common = apple.intersection(ms)
print(common)

al = apple.union(ms)
print(al)

o_apple = apple.difference(ms)
print(o_apple)
o_ms = ms.difference(apple)
print(o_ms)

u.difference(al)
```

```python
from bs4 import BeautifulSoup
from os import write
import csv
import requests
import pandas as pd

for i in range(10):
    data = []
    url = BeautifulSoup("http://www.nepalstock.com/main/floorsheet/index/0/","html.parser")
    soup = requests.get(url)
    lxml_code = BeautifulSoup(soup.text,'lxml')
    table_code = lxml_code.table
    tags = table_code.find_all("tr")

    for tag in tags:
        data.append(tag.text.split('\n'))
    data

f = open('NepseIndex.csv','w')
x = csv.writer(f)
for i in data:
    x.writerow(i)
f.close()
```

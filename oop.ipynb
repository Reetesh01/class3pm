```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return f"{self.x} {self.y}"
    
    def __add__(self,other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x,y)
    
p1 = Point(2,3)
p2 = Point(-1,3)
print(p1 + p2)
```

```python
class Ecommerce:
    def __init__(self,price):
        self.price = price
        print("This is init function",self.price)
              
    def __str__(self):
        print("this is str function",self.price)
        return str(self.price)
    
    def __add__(self,other):
        s = self.price + other.price
        print('this is add function',s)
        return Ecommerce(s)
    
p1 = Ecommerce(1000)
p2 = Ecommerce(2500)

print(p1+p2)
```

```python
class Ecommerce:
    def __init__(self,price):
        self.price = price
        print("This is init function",self.price)
              
    def __str__(self):
        print("this is str function",self.price)
        return str(self.price)
    
    def __add__(self,other):
        s = self.price + other.price
        print('this is add function',s)
        return Ecommerce(s)
    
p1 = Ecommerce(1000)
p2 = Ecommerce(2500)
p3 = Ecommerce(4000)

print(p1+p2+p3)
```

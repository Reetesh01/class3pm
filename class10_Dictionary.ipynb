```python
info = {'name':[],'ntc':[], 'ncell':[]}
n = int(input("enter n = "))
for i in range(n):
    name = input('enter name = ')
    ntc_phone = input("enter ntc number = ") 
    ncell_phone = input("enter ncell number  = ")
    
    info['name'].append(name)
    info['ntc'].append(ntc_phone)
    info['ncell'].append(ncell_phone)
    
print(info)
```

```python
info = []
n = int(input('enter n = '))
for i in range(n):
    name = input("enter name = ")
    age = int(input("enter age = "))
    add = input("enter address = ")
    
    data = {'name':name,'age':age,'add':add}
    
    info.append(data)

print(info,'\n')
```

```python
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter operator + - * / = ")

if o == "+":
    print(a+b)
    
elif o == "-":
    print(a-b)
    
elif o == "*":
    print(a*b)
    
elif o == "/":
    print(a/b)
    
else:
    print("invalid operator")
```

```python
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter operator + - * / = ")

if o == "+":
    print(a+b)
    
elif o == "-":
    print(a-b)
    
elif o == "*":
    print(a*b)
    
elif (o == "/") & (b!=0):
    print(a/b)
    
elif o == "/" and b==0:
    print("The value of b cannot be zero!!")
    
else:
    print("invalid operator")
```

```python
s=0
a=0
n=int(input("enter the value of n= "))
while a<n:
    x=int(input("enter value of x = "))
    s=s+x
    a+=1
print("The sum is =", s)
```

```python
s=str()
a=0
n=int(input("enter the value of n="))
while a<n:
    x=input("enter value of x =")
    s=s+x+"\n"
    a+=1
print("The sum is =", s)
```

```python
a = "Hello world i am python"
l = len(a)
i = 0
while i<l:
    if a[i] == " ":
        break
        
    print(a[i],end = "")
    
    i=i+1
```
